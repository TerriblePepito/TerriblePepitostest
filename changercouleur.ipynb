```python
from pathlib import Path
import re
# import tkinter as tk  # Removed: tkinter requires a graphical display, which is not available in Colab.
# from tkinter import filedialog # Removed

# In Google Colab, you need to specify the input directory path directly.

# --- OPTION 1: Using files uploaded directly to Colab session ---
# If you upload a folder named 'my_svgs' to /content/
# IN_DIR = Path("/content/my_svgs")

# --- OPTION 2: Using files from Google Drive ---
# Uncomment and run the following two lines to mount Google Drive first
from google.colab import drive
drive.mount('/content/drive')

# Then, specify the path to your SVG folder on Google Drive.
# For example, if your SVG files are in an 'input' folder directly under 'My Drive':
IN_DIR = Path("/content/drive/MyDrive/testinput") # <--- REMPLACEZ CECI PAR VOTRE VRAI CHEMIN

# Ensure the input directory exists
if not IN_DIR.exists():
    raise SystemExit(f"âŒ Input directory not found: {IN_DIR}. Please ensure this directory exists and contains your SVG files.")

# Define output directory. This will create an 'output' folder directly under '/content/drive/MyDrive'
OUT_DIR = Path("/content/drive/MyDrive/testoutput")
OUT_DIR.mkdir(exist_ok=True, parents=True) # Added parents=True for robustness

REPLACEMENTS = {
    "#95C11F": "#0000FF",
    "#FF995F": "#FFFF00",
}

def replace_colors(text):
    for old, new in REPLACEMENTS.items():
        text = re.sub(old, new, text, flags=re.IGNORECASE)
    return text

svg_files = list(IN_DIR.glob("*.svg"))
print(f"Found {len(svg_files)} SVG file(s).")

for svg in svg_files:
    try:
        text = svg.read_text(encoding="utf-8", errors="ignore")
        new_text = replace_colors(text)
        (OUT_DIR / svg.name).write_text(new_text, encoding="utf-8")
        print(f"âœ… Processed {svg.name}")
    except Exception as e:
        print(f"âŒ Failed to process {svg.name}: {e}")

print("ðŸŽ‰ Done")
```
