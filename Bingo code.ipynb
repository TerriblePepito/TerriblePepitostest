```python
import random

import random

class Bingo:
    def __init__(self):
        self.numeros_restants = list(range(1, 91))
        self.historique = []  # ordre exact des tirages

    def tirer(self):
        if not self.numeros_restants:
            return None  # fini

        n = random.choice(self.numeros_restants)
        self.numeros_restants.remove(n)
        self.historique.append(n)
        return n

    def historique_par_dizaine(self):
        # pr√©pare toutes les dizaines m√™me si vides
        result = {
            "1-9": [],
            "10-19": [],
            "20-29": [],
            "30-39": [],
            "40-49": [],
            "50-59": [],
            "60-69": [],
            "70-79": [],
            "80-89": [],
            "90": []
        }

        for n in self.historique:
            if n == 90:
                result["90"].append(n)
            elif n < 10:
                result["1-9"].append(n)
            else:
                debut = (n // 10) * 10
                result[f"{debut}-{debut+9}"].append(n)

        # trier l‚Äôint√©rieur de chaque dizaine
        for k in result:
            result[k].sort()

        return result

    def afficher(self):
        print("\n--- √âTAT DU JEU ---")
        print("Historique (ordre des tirages) :", self.historique)

        print("\nClass√© par dizaine :")
        classement = self.historique_par_dizaine()
        for dizaine, nums in classement.items():
            print(f"{dizaine:<7} : {nums}")

        print(f"\nRestants : {len(self.numeros_restants)}")
        print("-------------------\n")


# ---- Utilisation interactive ----
bingo = Bingo()

while True:
    cmd = input("Entr√©e = tirer | 'q' = quitter : ").strip().lower()
    if cmd == "q":
        break

    numero = bingo.tirer()
    if numero is None:
        print("‚úÖ Fin : tous les num√©ros ont √©t√© tir√©s.")
        bingo.afficher()
        break

    print(f"üé≤ Num√©ro tir√© : {numero}")
    bingo.afficher()









```
