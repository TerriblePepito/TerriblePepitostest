```python
rom pathlib import Path
import re

IN_DIR = Path("C:\Users\pierr\Documents\Pierre\Ingeone\code vor - jb\Test in")     # <-- mets le dossier source
OUT_DIR = Path("C:\Users\pierr\Documents\Pierre\Ingeone\code vor - jb\Test out")   # <-- dossier destination
OUT_DIR.mkdir(parents=True, exist_ok=True)

# Remplacements (insensibles √† la casse)
REPLACEMENTS = {
    "#95C11F": "#0000FF",  # green -> blue
    "#EA5B0C": "#FFFF00",  # orange -> yellow
}

def replace_colors_in_svg(svg_text: str) -> str:
    # remplace aussi les variantes en minuscules/majuscules
    for old, new in REPLACEMENTS.items():
        svg_text = re.sub(re.escape(old), new, svg_text, flags=re.IGNORECASE)
    return svg_text

for svg_path in IN_DIR.rglob("*.svg"):
    text = svg_path.read_text(encoding="utf-8", errors="ignore")
    new_text = replace_colors_in_svg(text)

    out_path = OUT_DIR / svg_path.relative_to(IN_DIR)
    out_path.parent.mkdir(parents=True, exist_ok=True)
    out_path.write_text(new_text, encoding="utf-8")

print("‚úÖ SVG trait√©s :", len(list(IN_DIR.rglob("*.svg"))))
print("üìÅ Sortie dans :", OUT_DIR.resolve())
```
